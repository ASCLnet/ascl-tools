```python
import requests
import github_config
import numpy as np
import pandas as pd
```

```python
#We nee to add year data to languages.csv

#must do api call to Github and API call to ASCL

#First, we'll start with Github
df = pd.read_csv("languages.csv")
df = df.drop(df.columns[0], axis=1)
print(len(df))

#get rid of all the abnormal rows
patternDel = "^\d*$"
df = df[df.Bytes.str.match(patternDel)]

print(len(df))

#get all the unique values of repos & author
newdf = df[['Repo','Author']].drop_duplicates()
print(len(newdf))

#set up HTTP call
session = requests.Session()
session.auth = (github_config.username, github_config.password)


#https://api.github.com/repos/{:owner}/{:repository}
#will return json -- get the created_at field
http_base = "https://api.github.com/repos/"

#test run
r = session.get(http_base+'0satoken/Eclairs')
json = r.json()
#print(json)
print(json['created_at'])
```

```python
#RUN THIS ONLY ONCE
arr = []

for index, row in newdf.iterrows():
    r = session.get(http_base+row['Author']+"/"+row['Repo'])
    arr.append(r.json()['created_at'])

print(arr)
```

```python
newdf['github_date'] = arr

#check one against a fresh API call to make sure it's correct
created = session.get(http_base+'Azeret/galIMF').json()['created_at']
print(created)
print(newdf[newdf['Repo'] == 'galIMF'])

#both are the same, so we're good
```
